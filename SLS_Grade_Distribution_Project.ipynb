What will my grade be in a future year based on my grades in my past classes?

```python
# Get inputs of number of classes, Class, Grade, and Prof, time they took the class and Future Time
import pandas as pd
import io


num_classes = int(input("Enter the number of stat classes you have taken: ")) # get number of classes
classnamelist = []
gradelist = []
proflist = []
timelist = []
seasonlist = []

for i in range(num_classes): # loop through number of classes
  classname = input("Enter a class: ") # get class name
  grade = input("Enter a grade: ") # get grade
  prof = input("Enter a professor: ") # get professor
  time = input("Enter what year you took the class: ") # get time
  season = input("Enter what season you took the class: ") # get season

  classnamelist.append(classname) # adding the classname to list
  gradelist.append(grade) # adding the grade to list
  proflist.append(prof) # adding the professor to list
  timelist.append(time) # adding the time to list
  seasonlist.append(season) # adding the season to list
```
